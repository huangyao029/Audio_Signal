```python
# -*- coding:utf-8 -*-
# Author : Younger Huang
# Date : 2023.11.07

import os, librosa
import numpy as np
import soundfile as sf
from scipy.linalg import hankel

def blms(x, d, N = 4, L = 4, mu = 0.1):
    nIters = min(len(x), len(d)) // L
    u = np.zeros(L + N - 1)
    w = np.zeros(N)
    e = np.zeros(nIters * L)
    for n in range(nIters):
        u[:-L] = u[L:]
        u[-L:] = x[n * L : (n + 1) * L]
        d_n = d[n*L : (n+1)*L]
        A = hankel(u[:L], u[-N:])
        e_n = d_n - np.dot(A, w)
        w = w + mu * np.dot(A.T, e_n) / L
        e[n * L : (n + 1) * L] = e_n
    return e

if __name__ == '__main__':
    x, sr = librosa.load(r'./res_6/x.wav', sr = 16000)
    v, sr = librosa.load(r'./res_6/d.wav', sr = 16000)
    e = blms(x, v, N = 256, L = 4, mu = 0.1)
    out_res_dir = './res_8_blms/'
    os.makedirs(out_res_dir, exist_ok = True)
    sf.write('%sx.wav'%(out_res_dir), x, sr, subtype = 'PCM_16')
    sf.write('%sd.wav'%(out_res_dir), v, sr, subtype = 'PCM_16')
    sf.write('%sblms.wav'%(out_res_dir), e, sr, subtype = 'PCM_16')
```

```python
import matplotlib.pyplot as plt

before_lms = './res_8_blms/d.wav'
after_lms = './res_8_blms/blms.wav'
before_lms_wav = sf.read(before_lms)[0]
after_lms_wav = sf.read(after_lms)[0]

plt.figure(figsize = (20, 20))
plt.subplot(411)
plt.plot(before_lms_wav)
plt.xlim((0, len(before_lms_wav)))
plt.title('Before BLMS')
plt.subplot(412)
plt.specgram(before_lms_wav)
plt.subplot(413)
plt.plot(after_lms_wav)
plt.xlim((0, len(after_lms_wav)))
plt.title('After BLMS')
plt.subplot(414)
plt.specgram(after_lms_wav)
plt.tight_layout(pad = 2, w_pad = 2)
plt.savefig('./before_after_blms_comp.png', dpi = 400)
plt.show()
```
